```
import matplotlib.pyplot as plt
import numpy as np

rule = [1, 1, 1, 0, 1, 0, 0, 0]

grid = np.random.randint(0, 2, size=(10, 10, 10), dtype=int)
for gen in range(200):
  print(f"Generation {gen}")
  fig,ax=plt.subplots(1,10,figsize=(30,20))
  for i,a in enumerate(ax.flatten()):
      a.imshow(grid[i,:,:], cmap='gray_r',)
      a.set_title(f"Column {i}")
  plt.show()

  for column in range(10):
    for rack in range(10):
      for host in range(10):
        if host == 0:
          right = 0
        else:
          right = grid[column][rack][host - 1]
        if host == 9:
          left = 0
        else:
          left = grid[column][rack][host + 1]
        cell = grid[column][rack][host]

        if(left == 0 and cell == 0 and right == 0):
          grid[column][rack][host] = rule[0]
        if(left == 0 and cell == 0 and right == 1):
          grid[column][rack][host] = rule[1]
        if(left == 0 and cell == 1 and right == 0):
          grid[column][rack][host] = rule[2]
        if(left == 0 and cell == 1 and right == 1):
          grid[column][rack][host] = rule[3]
        if(left == 1 and cell == 0 and right == 0):
          grid[column][rack][host] = rule[4]
        if(left == 1 and cell == 0 and right == 1):
          grid[column][rack][host] = rule[5]
        if(left == 1 and cell == 1 and right == 0):
          grid[column][rack][host] = rule[6]
        if(left == 1 and cell == 1 and right == 1):
          grid[column][rack][host] = rule[7]
```
